```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# define sentences
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good',
]

print(sent)
```

```python
voc_size = 10000
```

```python
one_hot_rep= [one_hot(words,voc_size)for words in sent]
one_hot_rep
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_max_length = 8
embedded_docs = pad_sequences(one_hot_rep,padding='post',maxlen=sent_max_length)
print(embedded_docs)
```

```python
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length = sent_max_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
